```
# Training corpus
corpus = [
    "<s> I love NLP </s>",
    "<s> I love deep learning </s>",
    "<s> deep learning is fun </s>"
]
# Simple whitespace tokenization
tokenized_corpus = [sentence.split() for sentence in corpus]

tokenized_corpus
from collections import defaultdict

unigram_counts = defaultdict(int)

for sentence in tokenized_corpus:
    for word in sentence:
        unigram_counts[word] += 1

print("Unigram Counts:")
for word, count in unigram_counts.items():
    print(word, ":", count)
bigram_counts = defaultdict(int)

for sentence in tokenized_corpus:
    for i in range(len(sentence) - 1):
        bigram = (sentence[i], sentence[i+1])
        bigram_counts[bigram] += 1

print("\nBigram Counts:")
for bigram, count in bigram_counts.items():
    print(bigram, ":", count)
bigram_probs = {}

for (w1, w2), count in bigram_counts.items():
    bigram_probs[(w1, w2)] = count / unigram_counts[w1]

print("\nBigram Probabilities (MLE):")
for bigram, prob in bigram_probs.items():
    print(bigram, ":", round(prob, 3))
def sentence_probability(sentence):
    tokens = sentence.split()
    prob = 1.0

    for i in range(len(tokens) - 1):
        bigram = (tokens[i], tokens[i+1])

        if bigram in bigram_probs:
            prob *= bigram_probs[bigram]
        else:
            return 0   # unseen bigram â†’ probability 0

    return prob
s1 = "<s> I love NLP </s>"
s2 = "<s> I love deep learning </s>"

prob_s1 = sentence_probability(s1)
prob_s2 = sentence_probability(s2)

print("Probability of S1:", prob_s1)
print("Probability of S2:", prob_s2)
if prob_s1 > prob_s2:
    print("\nModel prefers S1 because it has higher probability.")
elif prob_s2 > prob_s1:
    print("\nModel prefers S2 because it has higher probability.")
else:
    print("\nBoth sentences have equal probability.")
```
